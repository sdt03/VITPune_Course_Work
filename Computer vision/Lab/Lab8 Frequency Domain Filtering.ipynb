```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from numpy import dot, exp, mgrid, pi, ravel, square, uint8, zeros
```

```python
img = cv2.imread(r"/Users/shoumikdaterao/Desktop/TY Sem 5/Comp vision/Images//Serene.jpg",0)

f = np.fft.fft2(img)

fshift = np.fft.fftshift(f)

magnitude_spectrum = 20*np.log(np.abs(fshift))

phase_spectrum = np.angle(fshift)

plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])

plt.subplot(132),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])

plt.subplot(133),plt.imshow(phase_spectrum, cmap = 'gray')
plt.title('Phase Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('/Users/shoumikdaterao/Desktop/TY Sem 5/Comp vision/Images//Serene.jpg', cv2.IMREAD_GRAYSCALE)

f_transform = np.fft.fft2(img)
f_transform_shifted = np.fft.fftshift(f_transform)

# Create a high-pass filter mask
rows, cols = img.shape
center_row, center_col = rows // 2, cols // 2
high_pass_mask = np.ones((rows, cols))
high_pass_mask[center_row - 50:center_row + 50, center_col - 50:center_col + 50] = 0

# Apply the high-pass filter in frequency domain
f_transform_filtered = f_transform_shifted * high_pass_mask

# Perform inverse FFT to get the filtered image
filtered_image = np.fft.ifft2(np.fft.ifftshift(f_transform_filtered)).real

# Remove the 100x100 region from the filtered image
filtered_image[center_row - 50:center_row + 50, center_col - 50:center_col + 50] = 0

reconstructed_transform = np.fft.ifft2(np.fft.ifftshift(f_transform_shifted))
reconstructed_image = reconstructed_transform.real

plt.figure(figsize=(10, 20))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(np.abs(filtered_image), cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(reconstructed_image, cmap='gray')
plt.title('Reconstructed Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
