```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
image = cv2.imread("Img01.jpeg")

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title('Original Image')
```

```python
histr = cv2.calcHist([image],[0],None,[25],[0,255])
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Histogram for color scale picture')
plt.show()
```

```python
image=cv2.imread("Img01.jpeg")
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

blue_lower = np.array([100, 50, 50])
blue_upper = np.array([130, 255, 255])

mask = np.zeros(image.shape[:2], dtype="uint8")

blue_mask = cv2.inRange(hsv_image, blue_lower, blue_upper)

mask[blue_mask > 0] = 255

plt.figure(figsize=(12, 6))

for i, col in enumerate(color):
    histr = cv2.calcHist([image], [i], mask, [256], [0, 256])
    plt.plot(histr, color=col)
    plt.xlim([0, 256])

plt.title('Histograms for Blue Color Section')
plt.xlabel('Pixel Intensity')
plt.ylabel('Number of Pixels')

cv2.imshow("Masked Blue Section", cv2.bitwise_and(image, image, mask=mask))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image2=cv2.imread("Img03.jpeg")

image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(image2)
plt.title('Original Image')
```

```python
hist2 = cv2.calcHist([image2],[0],None,[255],[0,255])
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([image2],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Histogram for color scale picture')
plt.show()
```

```python
image2=cv2.imread("Img03.jpeg")
hsv_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2HSV)

brown_lower = np.array([10, 50, 50])
brown_upper = np.array([30, 255, 255])

brown_mask = cv2.inRange(hsv_image2, brown_lower, brown_upper)

brown_path = cv2.bitwise_and(image2, image2, mask=brown_mask)

remaining_mask = cv2.bitwise_not(brown_mask)

plt.figure(figsize=(12, 6))

for i, col in enumerate(color):
    histr = cv2.calcHist([image2], [i], remaining_mask, [256], [0, 256])
    plt.plot(histr, color=col)
    plt.xlim([0, 256])

plt.title('Histograms for Remaining Section (Excluding Brown Path)')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()

cv2.imshow("Original Image", image2)
cv2.imshow("Brown Path", brown_path)
cv2.imshow("Remaining Section", cv2.bitwise_and(image2, image2, mask=remaining_mask))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image_tire = cv2.imread("Tire01.jpeg", cv2.IMREAD_GRAYSCALE)
image_mri = cv2.imread("MRI01.jpeg", cv2.IMREAD_GRAYSCALE)

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
enhanced_tire = clahe.apply(image_tire)
enhanced_mri = clahe.apply(image_mri)

plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.imshow(image_tire, cmap='gray')
plt.title('Original Tire Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(enhanced_tire, cmap='gray')
plt.title('Enhanced Tire Image (CLAHE)')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(image_mri, cmap='gray')
plt.title('Original MRI Image')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(enhanced_mri, cmap='gray')
plt.title('Enhanced MRI Image (CLAHE)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
